### **Dictionary (Sözlük)**

1. Kapsayıcıdır
2. Sırasızdır ->  Listelerden farkıdır.
3. Değiştirilebilirdir.

##### **Dictionary Nedir**

Key'ler ve bu key'lerin karşılıklarının bir arada tutulduğu veri yapısıdır.
> Listelerde olduğu gibi index'leme yapılmaz.

##### **Dictionary Oluşturma**

```python
sozluk={"Reg":"regresyon modeli",
       "Loj":"lojistik regresyon",
       "Cart":"classification and reg"}
sozluk
```

```python
len(sozluk)
```

**{"key":"key'in karşılığı"}**

**Not**= Sözlüklerde key'ler sadece sabit veri yapılarından oluşabilir. List gibi yapılardan olamaz.

Sabit veri yapısı değiştirilemez demektir. Tuple'da buna dahildir.

t=("tuple",)->sozluk={t:"tuple'dan key olur"}

##### **Eleman Seçme İşlemleri**

```python
sozluk["Reg"]
```

```python
sozluk={"Reg":{"ASD":10,
               "XXX":20,
               "ZZZ":30},
       "Loj":{"ASD":10,
               "XXX":20,
               "ZZZ":30},
       "Cart":{"ASD":10,
               "XXX":20,
               "ZZZ":30}
       }
# sozluk icinde sozluk olusturduk,İç içe yapi.
sozluk["Reg"]["XXX"]
#Iç içe bir yapida elemana erişim
```

##### **Eleman Ekleme ve Değiştirme**

```python
sozluk={"Reg":"regresyon modeli",
       "Loj":"lojistik regresyon",
       "Cart":"classification and reg"}
# sozluk'e eleman ekleme
sozluk["Gbm"]="Gradient Boosting Mac"
sozluk
```

```python
# Reg Key'in karsiligini degistirme.
sozluk["Reg"]="Reg'in yeni karsiligi"
sozluk
```

**Not**= Reg key'i olmasaydı yeni key oluşturacaktı.

```python
t=("tuple",)# t adinda tuple olusturduk. Tek elemanlı
sozluk[t]="Tuple'dan key olusturuldu"
sozluk
```
